```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HR_Analytics.csv")
```

```python
df=df.dropna()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='AgeGroup', hue='Attrition')
plt.title("Attrition Rate Across Age Groups")
plt.xticks(rotation=45)
plt.show()
```

**Q2**

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='JobLevel', y='MonthlyIncome')
plt.title("Monthly Income Across Job levels")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x="YearsAtCompany", y="MonthlyIncome")
plt.title("Years at Company Vs Monthly Income")
plt.show()
```

```python
mean_income= np.mean(df['MonthlyIncome'])
median_income= np.median(df['MonthlyIncome'])
std_income= np.std(df['MonthlyIncome'])
print(mean_income, median_income, std_income)
```

```python
above_avg = np.sum(df['MonthlyIncome']>mean_income)/len(df)*100
above_avg
```
